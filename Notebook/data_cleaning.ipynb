```python
#Data Cleaning and Transformation
df_clean=df.copy()
```

```python
df_clean=df_clean.drop_duplicates()
```

```python
df_clean.isnull().sum()
```

```python
df_clean['Age_Group'] = pd.cut(
    df_clean['age'],
    bins=[0, 30, 45, 60, 100],
    labels=['Young', 'Middle-Aged', 'Senior', 'Elder']
)
```

```python
df_clean['Tenure_Group'] = pd.cut(
    df_clean['tenure'],
    bins=[-1, 2, 5, 10],
    labels=['New', 'Mid-Term', 'Long-Term']
)
```

```python
df_clean['Balance_Salary_Ratio'] = df_clean['balance'] / (df_clean['estimated_salary'] + 1)
```
